# Gamma Law Simulation by Rejection Sampling

This notebook illustrates a rejection-sampling approach for simulating Gamma-distributed random variables.
Before implementing the method itself, we first perform a simple sanity check on NumPy's pseudo-random number generator.

```python
import numpy as np
import matplotlib.pyplot as plt
```

We begin by generating uniform random variables with NumPy.
Before applying the rejection method, we visually check that the pseudo-random numbers $(X_1, \ldots, X_n)$ appear uniformly distributed on $[0,1]$ and that no obvious dependence is visible between two consecutive draws $(X_k, X_{k+1})$ for $k = 1, \ldots, n-1$. To do so, we plot both the histogram of the sample and the scatter plot of adjacent pairs. The latter should cover the unit square $[0,1] \times [0,1]$ as evenly as possible.

The figure below is obtained with a sample of size $n = 10{,}000$.

```python
rng = np.random.default_rng()
sample = rng.random(10000)
fig = plt.figure(figsize=(10, 5))
a1 = fig.add_subplot(121)
a1.scatter(sample[:-1], sample[1:], s=0.5, alpha=0.5)
a2 = fig.add_subplot(122)
a2.hist(sample, bins=20, alpha=0.2)
plt.show()
```

These diagnostics are qualitative, but they provide a useful first check before moving on to Gamma-law simulation.
